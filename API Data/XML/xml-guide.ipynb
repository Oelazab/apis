# XML Files in Python - Comprehensive Guide

## What is XML?

**XML (eXtensible Markup Language)** is a structured, hierarchical markup language designed to store and transport data in a human- and machine-readable format. Unlike JSON, XML supports attributes, namespaces, comments, and mixed content, making it the backbone of formats like HTML, SVG, RSS, SOAP, and many enterprise data systems.

## XML Structure Anatomy

```xml
<?xml version="1.0" encoding="UTF-8"?>   <!-- Declaration -->
<!-- This is a comment -->
<company name="TechCorp">                 <!-- Root element with attribute -->
    <employees>
        <employee id="1">                 <!-- Child element with attribute -->
            <name>Alice</name>            <!-- Text node -->
            <age>30</age>
            <skills>
                <skill>Python</skill>
                <skill>SQL</skill>
            </skills>
        </employee>
    </employees>
</company>
```
---
| Term | Description |
|---|---|
| **Declaration** | `<?xml version="1.0"?>` — optional header |
| **Element** | A tag pair: `<name>Alice</name>` |
| **Root Element** | The single top-level element wrapping everything |
| **Attribute** | Key-value inside a tag: `id="1"` |
| **Text Node** | The text content inside an element |
| **CDATA** | Raw unescaped text block: `<![CDATA[...]]>` |
| **Namespace** | Prefix to avoid tag name conflicts: `xmlns:ns=` |
| **Comment** | `<!-- comment -->` |

## 1. `xml.etree.ElementTree` - The Standard Library Workhorse

### Parsing XML from a String

````python
import xml.etree.ElementTree as ET
````

````python
xml_data = """
<library>
    <book id="1" genre="fiction">
        <title>The Great Gatsby</title>
        <author>F. Scott Fitzgerald</author>
        <year>1925</year>
        <price>12.99</price>
    </book>
    <book id="2" genre="science">
        <title>A Brief History of Time</title>
        <author>Stephen Hawking</author>
        <year>1988</year>
        <price>15.50</price>
    </book>
</library>
"""
````

````python
root = ET.fromstring(xml_data)
````

````python
print(root.tag)               # library
print(len(root))              # 2  (number of child elements)
````

````python
# Access first book
book = root[0]
print(book.tag)               # book
print(book.attrib)            # {'id': '1', 'genre': 'fiction'}
print(book.get("id"))         # 1
print(book.find("title").text)  # The Great Gatsby
````

### Parsing XML from a File

````python
import xml.etree.ElementTree as ET
````

````python
# Parse directly from file path
tree = ET.parse("library.xml")
root = tree.getroot()

print(root.tag)               # library
````

## 2. Navigating the XML Tree

### Iterating Children

````python
import xml.etree.ElementTree as ET

tree = ET.parse("library.xml")
root = tree.getroot()
````

````python
# Iterate direct children
for book in root:
    print(book.tag, book.attrib)
````

````python
# Iterate all descendants (any depth)
for element in root.iter():
    print(element.tag, "->", element.text)
````

````python
# Iterate specific tag name at any depth
for title in root.iter("title"):
    print(title.text)
````

### `find()` vs `findall()` vs `iter()`

````python
# find() — returns the FIRST matching element or None
author = root.find("book/author")
print(author.text)   # F. Scott Fitzgerald
````

````python
# findall() — returns a LIST of all direct/path matches
books = root.findall("book")
print(len(books))    # 2
````

````python
# Nested path search
titles = root.findall("book/title")
for t in titles:
    print(t.text)
````

````python
# iter() — recursive generator over all matching tags
for price in root.iter("price"):
    print(float(price.text))
````

### Accessing Attributes

````python
for book in root.findall("book"):
    book_id  = book.get("id")           # Returns None if missing
    genre    = book.get("genre", "N/A") # Default fallback value
    title    = book.find("title").text
    print(f"[{book_id}] {title} ({genre})")
````

## 4. Creating XML from Scratch

````python
import xml.etree.ElementTree as ET

# Create root element
root = ET.Element("catalog")
root.set("version", "1.0")
````

````python
# Add child elements
products = ET.SubElement(root, "products")
````

````python
# First product
p1 = ET.SubElement(products, "product")
p1.set("id", "101")
ET.SubElement(p1, "name").text   = "Laptop"
ET.SubElement(p1, "price").text  = "999.99"
ET.SubElement(p1, "stock").text  = "50"

# Second product
p2 = ET.SubElement(products, "product")
p2.set("id", "102")
ET.SubElement(p2, "name").text   = "Mouse"
ET.SubElement(p2, "price").text  = "29.99"
ET.SubElement(p2, "stock").text  = "200"
````

````python
# Build the tree and write to file
tree = ET.ElementTree(root)
ET.indent(tree, space="    ")      # Pretty-print (Python 3.9+)
tree.write("catalog.xml", encoding="utf-8", xml_declaration=True)
````

**Output `catalog.xml`:**
```xml
<?xml version='1.0' encoding='utf-8'?>

    <products>
        <product id="101">
            <name>Laptop</name>
            <price>999.99</price>
            <stock>50</stock>
        </product>
        <product id="102">
            <name>Mouse</name>
            <price>29.99</price>
            <stock>200</stock>
        </product>
    </products>
</catalog>
```

## 5. Modifying Existing XML

````python
import xml.etree.ElementTree as ET

tree = ET.parse("catalog.xml")
root = tree.getroot()
````

````python
# --- Update text content ---
for product in root.iter("product"):
    if product.get("id") == "101":
        product.find("price").text = "899.99"   # Price drop!
````

````python
# --- Update an attribute ---
for product in root.iter("product"):
    product.set("currency", "USD")
````

````python
# --- Add a new child element ---
for product in root.iter("product"):
    if product.find("discount") is None:
        ET.SubElement(product, "discount").text = "0"
````

````python
# --- Remove an element ---
products = root.find("products")
for product in products.findall("product"):
    if product.find("stock").text == "0":
        products.remove(product)   # Remove out-of-stock items
````

````python
# Save changes
ET.indent(tree, space="    ")
tree.write("catalog.xml", encoding="utf-8", xml_declaration=True)
````

````python
import xml.etree.ElementTree as ET
import xml.dom.minidom

tree = ET.parse("catalog.xml")
raw_string = ET.tostring(tree.getroot(), encoding="unicode")

dom = xml.dom.minidom.parseString(raw_string)
pretty_xml = dom.toprettyxml(indent="    ")

# Strip the extra declaration minidom adds
lines = pretty_xml.split("\n")[1:]   # Remove first <?xml?> line if duplicate
print("\n".join(lines))
````

## 7. Working with Namespaces

Namespaces prevent element name collisions in complex XML documents (e.g., SOAP, SVG).

````python
import xml.etree.ElementTree as ET

xml_ns = """
<root xmlns:dc="http://purl.org/dc/elements/1.1/"
      xmlns:ex="http://example.com/schema">
    <dc:title>My Document</dc:title>
    <ex:author>Alice</ex:author>
    <ex:date>2024-01-15</ex:date>
</root>
"""

root = ET.fromstring(xml_ns)
````

````python
# ✅ Correct: use Clark notation {uri}localname
title  = root.find("{http://purl.org/dc/elements/1.1/}title")
author = root.find("{http://example.com/schema}author")
print(title.text)   # My Document
print(author.text)  # Alice
````

````python
# Register namespace prefixes to control output
ET.register_namespace("dc", "http://purl.org/dc/elements/1.1/")
ET.register_namespace("ex", "http://example.com/schema")

# Define namespaces map for cleaner find() calls
ns = {
    "dc": "http://purl.org/dc/elements/1.1/",
    "ex": "http://example.com/schema"
}
````

````python
title = root.find("dc:title", ns)
print(title.text)   # My Document
````

## 8. SAX Parser - Streaming Large XML Files

For large XML files (hundreds of MB / GB), loading the entire tree into memory is impractical. SAX (Simple API for XML) parses **event-by-event**, keeping memory usage minimal.

````python
import xml.sax

class BookHandler(xml.sax.ContentHandler):
    def __init__(self):
        self.current_element = ""
        self.current_book = {}
        self.books = []

    def startElement(self, name, attrs):
        self.current_element = name
        if name == "book":
            self.current_book = {"id": attrs.get("id", "")}

    def characters(self, content):
        content = content.strip()
        if content and self.current_element in ("title", "author", "year"):
            self.current_book[self.current_element] = content

    def endElement(self, name):
        if name == "book" and self.current_book:
            self.books.append(self.current_book)
        self.current_element = ""
````

````python
# Run the parser
handler = BookHandler()
xml.sax.parse("library.xml", handler)

for book in handler.books:
    print(book)
````

````python
from lxml import etree

# Parse with full XPath 1.0 support
tree = etree.parse("library.xml")
root = tree.getroot()
````

````python
# Full XPath expressions
prices = root.xpath("//price[. > 13]/text()")
print(prices)   # ['15.50']
````

````python
# XPath with namespaces
ns = {"dc": "http://purl.org/dc/elements/1.1/"}
titles = root.xpath("//dc:title/text()", namespaces=ns)
````

````python
# Build XML programmatically
person = etree.Element("person", age="30")
etree.SubElement(person, "name").text = "Alice"
etree.SubElement(person, "email").text = "alice@example.com"

print(etree.tostring(person, pretty_print=True).decode())
````

````python
# Validate against an XML Schema (XSD)
with open("schema.xsd", "rb") as f:
    schema = etree.XMLSchema(etree.parse(f))

is_valid = schema.validate(tree)
print("Valid:", is_valid)
if not is_valid:
    print(schema.error_log)
````

## 10. `xmltodict` - XML ↔ Dictionary Conversion

install library
```bash
pip install xmltodict
```

````python
import xmltodict
import json

xml_str = """
<order id="500">
    <customer>Alice</customer>
    <items>
        <item sku="A1"><name>Pen</name><qty>3</qty></item>
        <item sku="B2"><name>Notebook</name><qty>1</qty></item>
    </items>
    <total>8.75</total>
</order>
"""
````

````python
# XML → Python OrderedDict
data = xmltodict.parse(xml_str)
print(data["order"]["customer"])          # Alice
print(data["order"]["items"]["item"][0])  # {'@sku': 'A1', 'name': 'Pen', 'qty': '3'}
````

````python
# Access attributes via @key convention
print(data["order"]["@id"])   # 500
````

````python
# Convert to JSON seamlessly
json_str = json.dumps(data, indent=4)
print(json_str)
````

````python
# Dict → XML
back_to_xml = xmltodict.unparse(data, pretty=True)
print(back_to_xml)
````

## 11. Error Handling

````python
import xml.etree.ElementTree as ET

def safe_parse_file(filepath):
    try:
        tree = ET.parse(filepath)
        return tree.getroot()
    except FileNotFoundError:
        print(f"File not found: {filepath}")
    except ET.ParseError as e:
        print(f"Malformed XML: {e}")
    except PermissionError:
        print(f"Permission denied: {filepath}")
    return None

def safe_get(element, tag, default="N/A"):
    """Safely extract text from a child element."""
    child = element.find(tag)
    if child is not None and child.text:
        return child.text.strip()
    return default
````

````python
# Usage
root = safe_parse_file("data.xml")
if root:
    for item in root.findall("item"):
        name  = safe_get(item, "name")
        price = safe_get(item, "price", "0.00")
        print(f"{name}: ${price}")
````

## 12. XML to [CSV / DataFrame]

````python
import xml.etree.ElementTree as ET
import csv

tree = ET.parse("library.xml")
root = tree.getroot()
````

XML → CSV

````python
with open("books.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["id", "genre", "title", "author", "year", "price"])
    for book in root.findall("book"):
        writer.writerow([
            book.get("id"),
            book.get("genre"),
            book.findtext("title"),
            book.findtext("author"),
            book.findtext("year"),
            book.findtext("price"),
        ])
````

XML → Pandas DataFrame (pip install pandas)

````python
import pandas as pd

records = []
for book in root.findall("book"):
    records.append({
        "id":     book.get("id"),
        "title":  book.findtext("title"),
        "author": book.findtext("author"),
        "year":   int(book.findtext("year")),
        "price":  float(book.findtext("price")),
    })

df = pd.DataFrame(records)
print(df)
````

## 13. CDATA Sections

CDATA sections embed raw text (e.g., HTML, scripts) inside XML without escaping.

````python
# lxml supports CDATA natively
from lxml import etree
````

````python
root = etree.Element("content")
root.text = etree.CDATA("<h1>Hello <b>World</b></h1>")
````

````python
print(etree.tostring(root, pretty_print=True).decode())
````

## Quick Reference Summary

| Task | Tool / Method |
|---|---|
| Parse XML file | `ET.parse("file.xml")` |

| Find all matches | `root.findall("tag")` |

| Read attribute | `element.get("attr")` |

| Pretty-print | `ET.indent(tree)` (3.9+) or `minidom` |
| Namespaces | Clark notation `{uri}tag` or ns dict |
| Stream large files | `xml.sax` |
| Full XPath / XSLT | `lxml` |
| XML ↔ dict | `xmltodict` |
| XSD Validation | `lxml.etree.XMLSchema` |
