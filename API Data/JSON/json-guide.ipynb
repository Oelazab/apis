# JSON Files in Python - Comprehensive Guide

## What is JSON?

**JSON (JavaScript Object Notation)** is a lightweight, human-readable data interchange format. Despite its JavaScript origins, it's language-agnostic and is one of the most widely used formats for data storage, APIs, and configuration files.

## JSON Data Types & Python Equivalents

| JSON Type | Example | Python Equivalent |
|---|---|---|
| `string` | `"hello"` | `str` |
| `number` | `42`, `3.14` | `int`, `float` |
| `boolean` | `true`, `false` | `True`, `False` |
| `null` | `null` | `None` |
| `array` | `[1, 2, 3]` | `list` |
| `object` | `{"key": "value"}` | `dict` |

## The `json` Module

Python ships with a built-in `json` module — no installation needed.

```python
import json
```

It provides four core functions:

| Function | Direction | Target |
|---|---|---|
| `json.dumps()` | Python → JSON string | In memory |
| `json.loads()` | JSON string → Python | In memory |
| `json.dump()` | Python → JSON file | File I/O |
| `json.load()` | JSON file → Python | File I/O |

## 1. Serialization - `dumps()` & `dump()`
Serialization converts Python objects **into** JSON format.

### `json.dumps()` - To a String

````python
import json
````

````python
data = {
    "name": "Alice",
    "age": 30,
    "active": True,
    "scores": [95, 87, 91],
    "address": None
}
````

````python
# Basic conversion
json_string = json.dumps(data)
print(json_string)
# {"name": "Alice", "age": 30, "active": true, "scores": [95, 87, 91], "address": null}
````

````python
# Pretty-printed with indentation
pretty = json.dumps(data, indent=4)
print(pretty)
````

````python
# Sort keys alphabetically
sorted_json = json.dumps(data, indent=4, sort_keys=True)
````

````python
# Ensure ASCII-safe output (default True)
json.dumps({"city": "München"}, ensure_ascii=False)  # preserves Unicode
````

### `json.dump()` - To a File

````python
import json

data = {"product": "Laptop", "price": 999.99, "in_stock": True}

with open("product.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
````

## 2. Deserialization - `loads()` & `load()`

Deserialization converts JSON **back** into Python objects.

### `json.loads()` - From a String

````python
import json
````

````python
json_string = '{"name": "Bob", "age": 25, "active": false}'
````

````python
data = json.loads(json_string)
````

````python
print(data["name"])    # Bob
print(data["age"])     # 25
print(type(data))      # <class 'dict'>
````

### `json.load()` - From a File

````python
import json

with open("product.json", "r", encoding="utf-8") as f:
    data = json.load(f)

print(data["product"])   # Laptop
print(data["price"])     # 999.99
````

## 3. Working with Nested JSON

Real-world JSON is typically deeply nested. Here's how to navigate and manipulate it.

````python
import json

json_data = """
{
    "company": "TechCorp",
    "employees": [
        {
            "id": 1,
            "name": "Alice",
            "skills": ["Python", "SQL", "Docker"],
            "contact": {"email": "alice@tech.com", "phone": "555-0101"}
        },
        {
            "id": 2,
            "name": "Bob",
            "skills": ["JavaScript", "React"],
            "contact": {"email": "bob@tech.com", "phone": "555-0102"}
        }
    ]
}
"""
````

````python
data = json.loads(json_data)
````

````python
# Access nested values
print(data["company"])                          # TechCorp
print(data["employees"][0]["name"])             # Alice
print(data["employees"][0]["skills"][1])        # SQL
print(data["employees"][1]["contact"]["email"]) # bob@tech.com
````

````python
# Iterate over nested structures
for emp in data["employees"]:
    print(f"{emp['name']}: {', '.join(emp['skills'])}")
````

## 4. Modifying JSON Data

````python
import json

json_data = """
{
    "company": "TechCorp",
    "employees": [
        {
            "id": 1,
            "name": "Alice",
            "skills": ["Python", "SQL", "Docker"],
            "contact": {"email": "alice@tech.com", "phone": "555-0101"}
        },
        {
            "id": 2,
            "name": "Bob",
            "skills": ["JavaScript", "React"],
            "contact": {"email": "bob@tech.com", "phone": "555-0102"}
        }
    ]
}
"""
data = json.loads(json_data)

# Save 
with open("data.json", "w") as f:
    json.dump(data, f, indent=4)
````

````python
with open("data.json", "r") as f:
    data = json.load(f)

# Add a new field
data["version"] = "2.0"

# Update an existing field
data["employees"][0]["name"] = "Alice Smith"

# Remove a field
del data["employees"][1]["contact"]["phone"]

# Append to a list
data["employees"].append({"id": 3, "name": "Charlie"})

# Save changes back
with open("data.json", "w") as f:
    json.dump(data, f, indent=4)
````

## 5. Error Handling

Always wrap JSON operations in try/except blocks in production code.

````python
import json
````

````python
# Handle malformed JSON
def safe_parse(json_string):
    try:
        return json.loads(json_string)
    except json.JSONDecodeError as e:
        print(f"JSON parse error: {e.msg}")
        print(f"  At line {e.lineno}, column {e.colno}")
        return None
````

````python
# Handle file errors
def load_config(filepath):
    try:
        with open(filepath, "r", encoding="utf-8") as f:
            return json.load(f)
    except FileNotFoundError:
        print(f"File not found: {filepath}")
    except json.JSONDecodeError as e:
        print(f"Invalid JSON in {filepath}: {e}")
    except PermissionError:
        print(f"Permission denied: {filepath}")
    return {}
````

````python
# Test it
result = safe_parse('{"name": "Alice", "age":}')  # Malformed!
# JSON parse error: Expecting value
````

## 6. Serializing Custom Objects

By default, the `json` module can't serialize custom class instances. You have two solutions.

### Option A - Custom Encoder

````python
import json
from datetime import datetime, date

class CustomEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, (datetime, date)):
            return obj.isoformat()   # "2024-01-15T10:30:00"
        if hasattr(obj, "__dict__"):
            return obj.__dict__      # Serialize any object's attributes
        return super().default(obj)

# Usage
data = {
    "event": "Launch",
    "timestamp": datetime.now(),
    "date": date.today()
}

json_string = json.dumps(data, cls=CustomEncoder, indent=4)
print(json_string)
````

### Option B - `default` Parameter

````python
import json
from datetime import datetime

def serialize(obj):
    if isinstance(obj, datetime):
        return {"__type__": "datetime", "value": obj.isoformat()}
    raise TypeError(f"Type {type(obj)} not serializable")

data = {"created_at": datetime.now()}
json.dumps(data, default=serialize, indent=4)
````

### Custom Decoder

````python
import json
from datetime import datetime

def decode_datetime(dct):
    for key, value in dct.items():
        if isinstance(value, str):
            try:
                dct[key] = datetime.fromisoformat(value)
            except ValueError:
                pass
    return dct

json_str = '{"name": "Event", "timestamp": "2024-01-15T10:30:00"}'
data = json.loads(json_str, object_hook=decode_datetime)
print(type(data["timestamp"]))  # <class 'datetime.datetime'>
````

## 7. JSON with APIs (Requests Library)

````python
import ijson
import urllib.request

# Public API that returns a JSON array of objects
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.geojson"

with urllib.request.urlopen(url) as response:
    i = 0
    # response is a file-like object → perfect for ijson streaming
    # Prefix becomes "features.item" because the array is inside "features"
    for quake in ijson.items(response, "features.item"):
        print(quake["properties"]["mag"], quake["properties"]["place"])
        i += 1
        if i > 10:
            break
````

## 9. JSON Schema Validation

Validate that JSON matches an expected structure using `jsonschema`.
```bash
pip install jsonschema
```

````python
import json
import jsonschema
````

````python
schema = {
    "type": "object",
    "required": ["name", "age", "email"],
    "properties": {
        "name":  {"type": "string", "minLength": 1},
        "age":   {"type": "integer", "minimum": 0, "maximum": 130},
        "email": {"type": "string", "format": "email"}
    },
    "additionalProperties": False
}
````

````python
def validate_user(data):
    try:
        jsonschema.validate(instance=data, schema=schema)
        print("Valid!")
    except jsonschema.ValidationError as e:
        print(f"Validation failed: {e.message}")

validate_user({"name": "Alice", "age": 30, "email": "alice@example.com"})  # Valid!
validate_user({"name": "Bob", "age": -5})   # Validation failed: -5 < 0
````

## 10. Useful Patterns & Tips

### Merging JSON Objects

````python
base = {"theme": "dark", "lang": "en", "timeout": 30}
override = {"lang": "fr", "timeout": 60}
````

````python
# Shallow merge (Python 3.9+)
merged = base | override
merged
````

````python
# Deep merge for nested dicts
def deep_merge(base, override):
    result = base.copy()
    for key, val in override.items():
        if key in result and isinstance(result[key], dict) and isinstance(val, dict):
            result[key] = deep_merge(result[key], val)
        else:
            result[key] = val
    return result

deep_merge(base, override)
````

### Safe Key Access with `.get()`

````python
data = {"user": {"name": "Alice"}}
````

````python
# Risky - raises KeyError if missing
try:
    name = data["user"]["address"]["city"]
except KeyError as e:
    print(f" error: {e.msg}")
````

````python
# Safe — returns None or a default
city = data.get("user", {}).get("address", {}).get("city", "Unknown")
````

### Pretty-Print Any JSON

````python
import json

def pprint_json(data):
    print(json.dumps(data, indent=4, ensure_ascii=False, sort_keys=True))

pprint_json(data)
````

### JSON ↔ Dataclass (Python 3.7+)

````python
import json
from dataclasses import dataclass, asdict

@dataclass
class User:
    name: str
    age: int
    email: str

user = User("Alice", 30, "alice@example.com")
````

````python
# To JSON
json_str = json.dumps(asdict(user), indent=4)
print(json_str)
````

````python
# From JSON
raw = json.loads(json_str)
restored = User(**raw)
print(restored)
````
